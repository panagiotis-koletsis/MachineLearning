```python
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_california_housing
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from scipy.stats import uniform
from scipy.stats import randint
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import RandomizedSearchCV, KFold
from sklearn.metrics import accuracy_score
import torch
```

```python
data = pd.read_csv(r'D:\μαθήματα σχολής\------------------------πτυχιακη\datasets\fetalHealth\fetal_health.csv')
```

```python
X = data.drop('fetal_health', axis=1)
y = data['fetal_health']
X_train, X_test, y_train, y_test = train_test_split(
X,y , random_state=51,test_size=0.3, shuffle=True)
```

```python
scaler = StandardScaler()
X_train_Sscaled = scaler.fit_transform(X_train)
X_test_Sscaled = scaler.transform(X_test)
```

```python
scaler = MinMaxScaler()
X_train_Mscaled = scaler.fit_transform(X_train)
X_test_Mscaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
# Create a logistic regression model
model = LogisticRegression()
# Define the hyperparameter distributions for RandomizedSearchCV
param_dist = {
    'C': uniform(0.1, 100.0),  
    'max_iter': [400,800,1200], 
}
# Create the RandomizedSearchCV object with 5-fold cross-validation
random_search = RandomizedSearchCV(estimator=model, param_distributions=param_dist, n_iter=10, scoring='accuracy', cv=5, random_state=42)
# Fit the model to the training data
random_search.fit(X_train_Mscaled, y_train)
# Print the best hyperparameters
print("Best hyperparameters:", random_search.best_params_)
# Access the best model
best_model_random = random_search.best_estimator_
# Make predictions on the test set
y_pred = best_model_random.predict(X_test_Mscaled)
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.tree import DecisionTreeClassifier
# Define the parameter grid to search
param_dist = {
    'max_depth': randint(1, 20),  # Example range for max_depth
    'min_samples_split': randint(2, 20),
    'min_samples_leaf': randint(1, 20),
    'max_features': ['auto', 'sqrt', 'log2', None],
    'splitter': ['best', 'random'],
}
# Create a DecisionTreeClassifier with 3 classes
clf = DecisionTreeClassifier(random_state=0)
# Use RandomizedSearchCV to perform random search
random_search = RandomizedSearchCV(clf, param_distributions=param_dist, n_iter=10, cv=5, scoring='accuracy', random_state=0)
# Fit the model to the data
random_search.fit(X_train, y_train)
# Get the best estimator and print the best parameters
best_clf = random_search.best_estimator_
print("Best parameters found: ", random_search.best_params_)
# Predict using the best model
y_pred = best_clf.predict(X_test)
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier
# Define the parameter grid to search
param_dist = {
    'n_estimators': randint(50, 200),
    'max_depth': randint(5, 30),
    'min_samples_split': randint(2, 20),
    'min_samples_leaf': randint(1, 20),
    'max_features': ['auto', 'sqrt', 'log2'],
}
# Create a RandomForestClassifier
clf = RandomForestClassifier(random_state=1)
# Use RandomizedSearchCV to perform random search
random_search = RandomizedSearchCV(
    clf, param_distributions=param_dist, n_iter=10, cv=5, scoring='accuracy', random_state=1
)
# Fit the model to the data
random_search.fit(X_train, y_train)
# Get the best estimator and print the best parameters
best_clf = random_search.best_estimator_
print("Best parameters found: ", random_search.best_params_)
# Perform cross-validation with the best model
cv_scores = cross_val_score(best_clf, X_train, y_train, cv=5)
print("Cross-validation scores:", cv_scores)
# Predict using the best model
y_pred = best_clf.predict(X_test)
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
# Define the parameter grid to search
param_dist = {
    'n_estimators': randint(50, 200),
    'max_depth': randint(5, 30),
    'min_samples_split': randint(2, 20),
    'min_samples_leaf': randint(1, 20),
}
# Create an ExtraTreesClassifier
clf = ExtraTreesClassifier(random_state=1)
# Use RandomizedSearchCV to perform random search
random_search = RandomizedSearchCV(
    clf, param_distributions=param_dist, n_iter=10, cv=5, scoring='accuracy', random_state=1
)
# Fit the model to the data
random_search.fit(X_train, y_train)
# Get the best estimator and print the best parameters
best_clf = random_search.best_estimator_
print("Best parameters found: ", random_search.best_params_)
# Perform cross-validation with the best model
cv_scores = cross_val_score(best_clf, X_train, y_train, cv=5)
print("Cross-validation scores:", cv_scores)
# Predict using the best model
y_pred = best_clf.predict(X_test)
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
